### Positional-Only and keyword-only Arguments

```python
def pos_only(x,/,y):
    return x+y
print(pos_only(2,y=4)) # x must be positonal
```

```python
def key_only(*,x,y):
    return x*y
print(key_only(x=4,y=8)) # 'x' and 'y' must be defined with key
```

### Default argument

```python
def salom(name="Mehmon"):
    print(f'Hello {name}!')
print(salom('Abdulla'))
print(salom())
```

```python
def power(base,degree=2):
    return base**degree
print(power(4))
print(power(3,8))
```

### Arbitrary Arguments

```python
# Arbitrary positiona argument (*args)
def sum_all(*args):
    return sum(args)
print(sum_all(1,2,3,456,9))
```

```python
# Arbitrary keyword argument (**kwargs)
def print_info(**kwargs):
    print(type(kwargs))
```
