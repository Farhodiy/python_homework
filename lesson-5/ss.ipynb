### Comprehensions
- list comprehension
- dictionary comp
- set comp
- generator expressions

```python
numbers = [2,4,5,6]
asquares = [num**2 for num in numbers]
print(asquares)
```

```python
a_evensq=[val**2 for val in numbers if val%2==0]
print(a_evensq)
```

```python
a_print=[print(val) for val in numbers if val%2==0]
print(a_print)
```

```python
cordinates=[(x,y) for x in range(3) for y in range(3) ]
print(cordinates)
```

### Dictionary comprehension


```python
keys=['a','b','c','d']
values=[1,2,3,4]
d={k:v for (k,v) in zip(keys,values)}
print(d)
```

```python
Numbers = [2,4,5,6]
squaress={num:num**2 for num in Numbers}
print(squaress)
```

```python
matrix=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
dic={pos:{num:num**2 for num in row } for pos, row in enumerate(matrix) }
print(dic)
```
